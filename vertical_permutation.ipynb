```python
from common.alhabets import ALHABET

def batch_text(text: str, size: int = 3) -> list[str]:    
    iter_text = (char for char in text if char in ALHABET.keys())
    batches = []
    for _ in range(len(text) // size):
        batch = ""
        for __ in range(size):
            try:
                char = next(iter_text)
                batch += char
            except StopIteration:
                batch += "X"    
        batches.append(batch)

    # Clean
    res = []
    for i, batch in enumerate(batches[::]):
        if batch.replace("X", "") != "":
            res.append(batch)


    return res
```
